## Read data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('salary.csv')
df.head()
```

Using aliases for various columns of the dataframe

Gender Can be modified from F and M to binary values so that wecan use it as a colouring feature

```python
x=df['Years with Company']
y=df['Salary']
z=df['Age']

df['Gender'] = df['Gender'].replace(['F'],1)
df['Gender'] = df['Gender'].replace(['M'],0)
w=df['Gender']
```

```python
count = df['WorkType'].value_counts() 

WorkType=df.WorkType.unique()
print(WorkType)
```

Calculating Mean age according to WorkType and Visualizing it as a bar chart

Note: We can do the same with median as well

```python
new_df=df.groupby('WorkType').mean()
print(new_df['Age'])
columns=new_df.columns
print(columns)
```

```python
plt.bar(WorkType,new_df['Age'])
plt.xlabel("WorkType")
plt.ylabel("Mean Age")
plt.show()
```

Seeing the worktypes of different people as a pie-chart

As you can guess this is not very helpful as it does not give us much insight

```python
fig = plt.figure(figsize =(10, 7)) 
plt.pie(z, labels = df['WorkType']) 
  
# show plot 
plt.show() 
```

Now let's try to get some insights out of the data, for example how is the salary correlated to various other factors.

```python
corr1=x.corr(y)
corr2=z.corr(y)
corr3=w.corr(y)
print("Correlation between Years with company and Salary ",corr1)
print("Correlation between Age and Salary ",corr2)
print("Correlation between Gender and Salary ",corr3)
```

```python
print(df.head())
df.hist()
plt.show()
```

Now let's visualize the relation between Age and Salary through a scatterplot, Simple enough?


```python
plt.scatter(z,y)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

Now I would like to see how the same distribution with Genders highlighted. We can use c attribute for colours.

```python
plt.scatter(z,y,c=w)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

Another idea to use an additional feature is to chamge the size of each point according to Years with Company.

This converts the Scatter Plot to a bubble chart.

```python
plt.scatter(z, y, s=x*10, alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

We can combine the two methods above to see a colourful bubble plot.

```python
plt.scatter(z, y, c=w, s=x*10, alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel("Years with company")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x,y,c=w)
plt.xlabel("Years with company")
plt.ylabel("Salary")
plt.show()
```

All these visualizations basically confirm our initial correlation calculations and show that the Salary positively correlates to Age and Years with Company but not very strongly. And we can also see the outliers clearly.
